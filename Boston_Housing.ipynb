Correlation Coeff

The correlation coefficient ranges from -1 to 1. If the value is close to 1, it means that there is a strong positive correlation between the two variables. When it is close to -1, the variables have a strong negative correlation.

````python
# measuring the linear relationships between the variables

correlation_matrix = boston.corr()
sns.heatmap(data=correlation_matrix, annot=True, cmap='vlag')
````

````python
# removing the upper triangular matrix

msk = np.zeros(correlation_matrix.shape, dtype=bool)
msk[np.triu_indices(len(msk))] = True
sns.heatmap(data=correlation_matrix, annot=True, mask=msk, cmap='vlag')
````

##LINEAR REGRESSION

Preparing the data for training the model

````python
boston
````

````python
X = boston.drop(['DIS', 'NOX', 'CHAS', 'TAX'], axis = 1)
X.drop(['MEDV'], axis = 1, inplace=True)
Y = boston['MEDV']
````

````python
X
````

````python
Y
````

````python
from sklearn.model_selection import train_test_split
````

````python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=45)
````

````python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lin_model = LinearRegression()
````

````python
lin_model.fit(X_train, Y_train)
````

````python
# model evaluation for training set
Y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, Y_train_predict)))
r2 = r2_score(Y_train, Y_train_predict)
# r2_score is basically the proportion of the variance in the dependent variable that is predictable from the independent variable.

print("The model performance for training set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# model evaluation for testing set
Y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, Y_test_predict)))
r2 = r2_score(Y_test, Y_test_predict)

print("The model performance for testing set")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
````

rmse refers to the root mean squared error of the predicted values.

r2_score (R-squared) is a statistical measure of how close the data are to the fitted regression line.

r2_score documentation: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html

````python
plt.plot(Y_test_predict,Y_test,'ro')
plt.plot([0,50],[0,50], 'g-')
plt.xlabel('predicted')
plt.ylabel('real')
````

##LOGISTIC REGRESSION

````python
x = boston.drop('CHAS', axis = 1)
y = boston['CHAS']
````

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 101)
````

````python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
````

````python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(random_state=101, max_iter=1800)
logmodel.fit(x_train,y_train)
````

````python
predictions = logmodel.predict(x_test)
from sklearn.metrics import classification_report,accuracy_score
````

````python
print(accuracy_score(y_test,predictions))
````

````python
print(classification_report(y_test, predictions))
````
